```python
m = 80
r = 0.5
g = -9.8
D = 1.29
A = 3.14 * r**2
```

```python
x = 100
v = 0
```

```python
def drag(v):
    return 0.5*D*A*v**2

def weight(m):
    return m*g

def model_change(x,v):
    dx = v
    dv = (weight(m) + drag(v)) / m
    return (dx, dv)
```

```python
import numpy as np
t_final = 6.0
step = .1
tt = np.arange(0,t_final,step)
val = (x,v)
```

```python
lista_vals = [val]

for _ in tt:
    delta_x, delta_v = model_change(*val)
    x += step * delta_x
    v += step * delta_v
    val = (x,v)
    lista_vals.append(val)
lista_vals = lista_vals[:-1]
```

```python
xv = list(zip(*lista_vals))
```
